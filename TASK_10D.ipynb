```python
import matplotlib.pyplot as plt

# Define the data
sizes = [35, 25, 25, 15]
mylabels = ['w', 'x', 'y', 'z']
myexplode = [0.2, 0, 0, 0]

# Plot the pie chart
# autopct='%1.1f%%' formats the percentage values displayed on each slice
plt.pie(sizes, labels=mylabels, explode=myexplode, shadow=True, autopct='%1.1f%%')

# Add a title
plt.title('Pie Chart')

# Display the chart
plt.show()
```
